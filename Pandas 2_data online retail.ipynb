```python
df1=df.copy()
```

```python
df1['Thang'] = pd.to_datetime(df['InvoiceDate']).dt.month
df1
```

```python
''' Tạo cột mới có tên quý –  ‘Previous’ nhận giá trị 1 nếu ngày lập hóa đơn nằm trong các tháng 1,2,3; 
nhận giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng 4,5,6; 
nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9;
nhận giá trị 4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12 '''
def quy(_month):
    if _month in [1,2,3]:
        return 1
    if _month in [4,5,6]:
        return 2
    if _month in [7,8,9]:
        return 3
    if _month in [10,11,12]:
        return 4
```

```python
df1['quy'] = df1['Thang'].apply(lambda x:quy(x))
df1
```

```python
#Tạo một cột mới có tên ‘Amount’ có giá trị bằng Quantity * UnitPrice
df1['Amount'] = df1['Quantity'] * df1['UnitPrice']
df1
```

```python
#Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 5% nếu là ‘France’ ngược lại là 0%
def calculate_discount(country, quarter):
    if country == 'United Kingdom' and quarter == 4:
        return 0.1
    elif country == 'France':
        return 0.05
    else:
        return 0
```

```python
df1['Discount'] = df1.apply(lambda row: 0.1 if row['Country'] == 'United Kingdom' and row['quy'] == 4 
                          else (0.05 if row['Country'] == 'France' else 0), axis=1)
df1
```

```python
# Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount.
df1['total'] = df1['Amount'] - (df1['Amount']*df1['Discount'])
df1
```
