## Tiến hành xử lý dữ liệu khuyết thiếu

```python
# Tiến hành loại bỏ dữ liệu khuyết thiếu, số lượng bản ghi còn lại là bao nhiêu %
```

```python
# Loại bỏ dữ liệu khuyết thiếu
df1 = df.dropna(axis=0)
```

```python
# % số lượng bản ghi còn lại
df1.shape[0]/df.shape[0]
```

## Tiến hành thay thế dữ liệu khuyết thiếu bằng cách sử dụng các giá trị thay thế: mode với các thuộc tính có giá trị rời rạc, median, nội suy với các biến có giá trị liên tục.

```python
#Sử lý giá trị khuyết thiếu bằng phương pháp nội suy
df1 = df.interpolate()
```

```python
df1.boxplot()
```

```python
# vẽ biểu đồ box plot cho MonthlyIncome
sns.boxplot(df1["MonthlyIncome"])
```

## Tiến hành xử lý dữ liệu ngoại lai

```python
# tính giá trị Q1 và Q3
Q1 = df1.quantile(0.20)
Q3 = df1.quantile(0.80)
# tính IQR 
IQR = Q3-Q1
# lọc dữ liệu ngoại lai
df2 = df1[~((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df2.shape[0]-df1.shape[0]
```

```python
df2.boxplot()
```

```python
df2.head()
```

```python
sns.kdeplot(data = df2['MonthlyIncome'])
```

## Chuẩn hoá dữ liệu

```python
# chuẩn hóa với minmax scaling
scaler = MinMaxScaler()

mms = scaler.fit_transform(pd.DataFrame(df2['MonthlyIncome']))
```

```python
sns.kdeplot(data = mms)
```

```python
# chuẩn hóa với robust scaling
scaler = RobustScaler()

rbs = scaler.fit_transform(pd.DataFrame(df2['MonthlyIncome']))
```

```python
sns.kdeplot(data = rbs)
```

```python
# chuẩn hóa với standard scaling
scaler = StandardScaler()

sc = scaler.fit_transform(pd.DataFrame(df2['MonthlyIncome']))
```

```python
sns.kdeplot(data = sc)
```
